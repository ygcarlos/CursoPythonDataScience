```python
import numpy as np
x =np.linspace(0,5,11)
y = x **2
```

```python
#Functional

plt.plot(x,y) #.plot é para exibir o grafico de acordo os valores definidos
#plt.show() pode ser usado aqui, funciona como print (apenas caso não utilize o Jupyter Notebook)
```

```python
plt.plot(x,y,'r') # o 'r' é o parâmetro para alterar a cor da linha (nesse caso, 'r' é vermelho)
plt.xlabel('X Label') #.xlabel é para acrescentar o t[itulo do eixo X
plt.ylabel('Y Label') #.ylabel é para acrescentar o título do eixo Y
plt.title('Title') # .title pra acescentar título ao gráfico em exibição
```

.subplot

```python
plt.subplot(1,2,1) #.subplot serve para organizar plots em uma grade nesse caso (1 linha, 2 colunas, plot número 1)
plt.plot(x,y,'r')

plt.subplot(1,2,2) # nesse caso (1 linha, 2 colunas, plot número 2)
#caso as subplot tenham o mesmo número, vai acontecer um processo de 'merge' e as duas linhas vão estar no mesmo gráfico
plt.plot(y,x,'b') # b de blue
```

#OO - Object Oriented

```python
fig = plt.figure() #.figure cria uma nova "figura", como se fosse um quadro em branco

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #o .add_axes([]) adiciona eixos a figura, detalhe, os valores são colocado dentro de uma lista ([])
# sendo o 0.1,0.1 representando o eixo da esquerda e de baixo e o 0.8,0.8 representando a largura (width) e altura (height), essa relação é usada para um quadro vazio.

axes.plot(x,y) #como aqui é um metodo de objeto orientado, é possivel utilizar o .plot diretamente 
axes.set_xlabel('X Label') #nessa ocasição de eixos, é necessario utilizar o .set_xlabel para que seja acrescentado o título, o mesmo vale para os exemplos abaixo
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.35]) #0.2 = 20% da esquerda, 0.5 = 50% de baixo, 0.4 = 40% do width, 0.35 = 35% do height

#no caso do axes2, alterar o valor da esquerda e baixo resultará em posicionamento diferente do grafico

axes1.plot(x,y)
axes1.set_title('Larger Plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```
