Plot Appearance

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x =np.linspace(0,5,11)
y = x **2
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='fuchsia',lw=3,alpha=0.5,linestyle='-',marker='o',markersize=15,markerfacecolor='blue',markeredgewidth=3,markeredgecolor='green')
#é possivel mudar a cor da linha através do argumento color utilizando um Hex Code ou o nome em inglês
#o argumento linewidth=x aumenta da grossura da linha, (pode se utilizar como lw=x)
#o argumento alpha=x altera a transparência
#o linestyle= altera o tipo da linha podendo ser '-', '--', ':', '-.' (pode se utilizar como utilizar ls=x)
#o marker=x acrescenta uma marcação as linhas
#o markersize=x altera o tamanho do marker
#o markerfacecolor= altera a cor de fundo do marker
#o markeredgewidth=x altera o tamanho da grossura do contorno do marker
#o markeredgecolor= altera a cor do contorno do marker
```
