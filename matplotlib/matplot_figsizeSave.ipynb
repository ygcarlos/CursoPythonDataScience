```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x =np.linspace(0,5,11)
y = x **2
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2) #.suplots() para gerar mais figuras, o parâmetro nrows=1 representa que seja exibido apenas uma linha de plots,
# ncols=2 representa que serão exibidos duas colunas de plots

#plt.tight_layout para que as numerações nos eixos crie um espaçamento para não ficar uma em cima da outra

#axes.plot(x,y) não funciona, nesse caso se utiliza o exemplo abaixo

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y) #também é possivel determinar os valores nos eixoes através de seleção de array, sendo 0 a primeira coluna e 1 a segunda coluna
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout() #é recomendado usar em todos os plots para que  não haja problema de espaçamento
```

#Figure Size e DPI

```python
fig = plt.figure(figsize=(3,2)) #o figsize representa o tamanha do quadro, o priemiro valor aumenta o tamanho horizontal 
#e o segundo valor aumenta o tamanho vertical

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
#fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2)) #também é possivel manipular o tamanho dentro de subplots

#axes[0].plot(x,y)

#axes[1].plot(y,x)

#plt.tight_layout()

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X ao Quadrado') #o label vai servir como etiqueta ao gráfico
ax.plot(x,x**3,label='X ao Cubo')

ax.legend() #.legend vai exibir as labels inclusas nos eixos
#também é possivél alterar a posição da legend com o parâmetro (loc=) que de acordo com o número, é colocado em determinada parte do gráfico
#o ideal é não acrescentar o parâmetro pois o 0 é atribuido automaticamente como a melhor posição disponivel no gráfico  
```
