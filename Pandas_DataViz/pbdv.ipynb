# Folhas de estilos

Matplotlib tem [style sheets](http://matplotlib.org/gallery.html#style_sheets) que você pode usar para deixar seus gráficos um pouco mais bonitos. Essas folhas de estilo incluem plot_bmh,plot_fivethirtyeight,plot_ggplot e muito mais. Eles basicamente criam um conjunto de regras de estilo que seus gráficos seguem. Eu recomendo usá-los, eles fazem com que todos os seus plots tenham a mesma aparência e se sintam mais profissionais. Você pode até criar o seu próprio se quiser que todos os gráficos da sua empresa tenham a mesma aparência (embora seja um pouco tedioso criar).

Aqui está como usá-los.

**Antes de plt.style.use() seus gráficos ficam assim:**

```python
df1['A'].hist()
```

Chame o estilo:

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

Agora seu plot ficará assim

```python
df1['A'].hist()
```

```python
plt.style.use('bmh')
df1['A'].hist()
```

```python
plt.style.use('dark_background')
df1['A'].hist()
```

```python
plt.style.use('fivethirtyeight')
df1['A'].hist()
```

```python
plt.style.use('ggplot')
```

Vamos ficar com o style ggplot e realmente mostrar como utilizar os recursos de plotagem integrados do pandas!

# Plot Types

Existem vários tipos de gráficos embutidos nos pandas, a maioria deles gráficos estatísticos por natureza:

* df.plot.area     
* df.plot.barh     
* df.plot.density  
* df.plot.hist    
* df.plot.scatter
* df.plot.bar      
* df.plot.box      
* df.plot.hexbin   
* df.plot.kde      
* df.plot.pie

Você também pode simplesmente chamar df.plot(kind='hist') ou substituir esse tipo de argumento por qualquer um dos termos-chave mostrados na lista acima (por exemplo, 'box','barh', etc.)

Vamos começar a passar por eles!

## Area

```python
df2.plot.area(alpha=0.4)
```

## Barplots

```python
df2.head()
```

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

## Histograma

```python
df1['A'].plot.hist(bins=50)
```

## Scatter Plots

```python
df1.plot.scatter(x='A',y='B')
```

Você pode usar 'c' para colorir com base em outro valor de coluna
Use cmap para indicar o mapa de cores a ser usado.
Para todos os mapas de cores, confira: http://matplotlib.org/users/colormaps.html

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```


Ou use s para indicar o tamanho com base em outra coluna. O parâmetro s precisa ser um array, não apenas o nome de uma coluna:

```python
df1.plot.scatter(x='A',y='B',s=df1['C']*200)
```

## Boxplots

```python
df2.plot.box()
```

## Hexagonal Bin Plot

Útil para dados bivariados, alternativa ao scatterplot:

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='Oranges')
```

## Kernel Density Estimation plot (KDE)

```python
df2['a'].plot.kde()
```

```python
df2.plot.density()
```

É isso! Espero que você possa ver por que esse método de plotagem será muito mais fácil de usar do que o matplotlib completo, ele equilibra a facilidade de uso com o controle sobre a figura. Muitas das chamadas de plotagem também aceitam argumentos adicionais do pai matplotlib plt.
