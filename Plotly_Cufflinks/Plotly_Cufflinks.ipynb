# Criando dados

```python
df = pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
```

```python
df.head()
```

```python
df2 = pd.DataFrame({'Category':['A','B','C'],'Values':[32,43,50]})
```

```python
df2
```

## Usando Cufflinks e iplot()

* scatter
* bar
* box
* spread
* ratio
* heatmap
* surface
* histogram
* bubble

```python
df.iplot()
```

## Scatter

```python
df.iplot(kind='scatter', x='A', y='B', mode='markers', size=20) #mode altera a visualização do gráfico
```

## Bar

```python
df2.iplot(kind='bar', x='Category', y='Values')
```

```python
df.count().iplot(kind='bar')
```

```python
df.sum().iplot(kind='bar')
```

## Boxplots

```python
df.iplot(kind='box')
```

## 3D Surface

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[500,400,300,200,100]})
```

```python
df3.iplot(kind='surface',colorscale='rdylbu')
```

## Spread

```python
df[['A','B']].iplot(kind='spread')
```

## Histogram

```python
df.iplot(kind='hist',bins=25)
#para selecionar um, grupo específico, utilize o df['A']
```

## Bubble

```python
df.iplot(kind='bubble',x='A',y='B',size='C')
```
