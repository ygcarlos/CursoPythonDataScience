# Categorical Plots

Agora vamos discutir o uso seaborn parar dar plot dados categóricos! Existem alguns tipos de plotagem principais para isso:

* factorplot
* boxplot
* violinplot
* stripplot
* swarmplot
* barplot
* countplot

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

### barplot

Mostra estimativas pontuais e intervalos de confiança como barras retangulares.

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
#o total de contas do público masculino é visivelmente maior que do público feminino
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
#o parâmetro estimator= é uma função de estimativa dentro de cada bin categórico
#exibindo os dados em estimativa de desvio padrão (standard)
```

### countplot

Isso é essencialmente o mesmo que o barplot, exceto que o estimador está contando explicitamente o número de ocorrências. É por isso que passamos apenas o valor x:

```python
sns.countplot(x='sex',data=tips)
```

```python
#contem mais pessoas do sexo masculino do que do sexo feminino
```

### boxplot

Um boxplot (ou box-and-whisker plot) mostra a distribuição de dados quantitativos de uma maneira que facilita as comparações entre variáveis ​​ou entre os níveis de uma variável categórica

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

### violinplot

Um violinplot desempenha um papel semelhante ao de um box and whisker plot. Ele mostra a distribuição de dados quantitativos em vários níveis de uma (ou mais) variáveis ​​categóricas de forma que essas distribuições possam ser comparadas. Ao contrário de um boxplot, no qual todos os componentes do gráfico correspondem a pontos de dados reais, o violinplot apresenta uma estimativa de densidade do kernel da distribuição subjacente.


```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
#violinplot é mais complexo de interpretar em comparação ao boxplot
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
#split une as condições para comparação lado a lado
```

### stripplot

O stripplot desenhará um gráfico de dispersão onde uma variável é categórica. Um stripplot pode ser desenhado sozinho, mas também é um bom complemento para um boxplot ou violiplot nos casos em que você deseja mostrar todas as observações junto com alguma representação da distribuição subjacente.

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False)

#por padrão, o jitter vem True, é mais facil visualizar, como no exemplo acima com o hue incluso
```

### swarmplot

O swarmplot é semelhante ao stripplot(), mas os pontos são ajustados (apenas ao longo do eixo categórico) para que não se sobreponham. Isso dá uma melhor representação da distribuição de valores, embora não seja tão dimensionado para um grande número de observações (tanto em termos da capacidade de mostrar todos os pontos quanto em termos de computação necessária para organizá-los).

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
#uma combinação de stripplot com violinplot, mostra a mesma distribuição exceto quanto é exibido os mesmos pontos
#ideal NÃO utilizar o swarmplot quando a estutura de dados for muito grande
```

### combinando violinplot com swarmplot

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
#é mais util para apresentação, mas apenas para análise de dados exploratórios
#para exibir dados para outra pessoa é mais válido utilizar boxplots, countplots ou barplots
```

### factorplot

factorplot é a forma mais geral de um gráfico categórico. Pode receber um parâmetro **kind** para ajustar o tipo de gráfico:

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```
