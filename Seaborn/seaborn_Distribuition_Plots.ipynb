### Distribuiton Plots

```python
import seaborn as sns #importando o seaborn
```

```python
%matplotlib inline
```

Seaborn vem com estrutura de dados embutidos

```python
tips = sns.load_dataset('tips') #carregando uma estrutura de um repositório online
```

```python
tips.head() #exibindo os 5 primeiros resultados
```

### distplot

```python
sns.distplot(tips['total_bill']) #distplot exibe a distribuição de uma definição univariada de observações
```

```python
sns.distplot(tips['total_bill'],kde=False, bins=30)

#kde = kernel density estimation, quando especificado como False, remove a projeção kde e exibe apenas o histograma
#histograma = gráfico de barras que demonstra uma distribuição de frequências
#o parâmetro bins=x especifica a quantidade de barras serão exibidas no gráfico (o número depende da estrutura de dados que será utilizado)
```

```python
#o gráfico aponta que a maior parte das contas estão entre 10 e 20 dólares
```

### jointplot

jointplot() permite que você combine dois distplots para dados bivariada. com sua escolha de qual **tipo(kind)** parâmetro você quer comparar:

* "scatter"
* "reg"
* "resid"
* "kde"
* "hex"

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
#no meio está o gráfico em scatter
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
# parâmetro kind para alterar o tipo do gráfico a ser exibido
```

```python
#no meio está o gráfico em hexágonos (kind='hex')
#quanto mais escuro o hexágono, maior a concentração de pessoas
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
#no meio está o gráfico em regressão (kind='reg')
#é semelhante ao scatter, porém, é acrescentado a linha de regressão

#regressão = é uma ferramenta estatística utilizada para o monitoramento de processos que possuem interferência conjunta de variáveis,
#utiliza duas ou mais variáveis que apresentam relação de dependência entre si.

#será mais discutido quando chegar ao tópico de machine learning.
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
#no meio está o gráfico em densidade (kind='kde')
#o menor circulo é onde tem mais concentração
```

### pairplot

pairplot vai emparelhar relacionamentos pares em um data frame inteiro (para as colunas numericas) e um argumento de matiz de cores (para colunas categóricas)

```python
sns.pairplot(tips)
```

```python
#pairplot é uma boa opção para visualizar dados rapidamente
```

```python
sns.pairplot(tips,hue='sex',palette='rocket')
```

```python
#com o parmâmetro (hue='') é possivel colorir alguma coluna categórica
#nesse caso a coluna selecionada foi 'sex' na qual o masculino ficou com uma cor e feminino ficou com outra cor criando um novo destaque
#o parâmetro palette='' altera o esquema de cores que vão ser utilizados na coluna especificada no hue
```

### rugplot

rugplots um conceito muito simples, eles apenas desenham um traço para cada ponto em uma distribuição univariada. Eles são o bloco de construção de um gráfico do KDE

```python
sns.rugplot(tips['total_bill'])
```

### kdeplot

kdeplots são [Kernel Density Estimation plots](https://pt.wikipedia.org/wiki/Estimativa_de_densidade_kernel). Esses KDE plots substituem cada observação por uma distribuição Gaussiana (Normal) centrada em torno desse valor.

```python
sns.kdeplot(tips['total_bill'])
```
