### Distribuiton Plots

```python
import seaborn as sns #importando o seaborn
```

```python
%matplotlib inline
```

Seaborn vem com estrutura de dados embutidos

```python
tips = sns.load_dataset('tips') #carregando uma estrutura de um reposit√≥rio online
```

```python
tips.head() #exibindo os 5 primeiros resultados
```

### distplot

```python
sns.distplot(tips['total_bill']) #distplot exibe a distribui√ß√£o de uma defini√ß√£o univariada de observa√ß√µes
```

```python
sns.distplot(tips['total_bill'],kde=False, bins=30)

#kde = kernel density estimation, quando especificado como False, remove a proje√ß√£o kde e exibe apenas o histograma
#histograma = gr√°fico de barras que demonstra uma distribui√ß√£o de frequ√™ncias
#o par√¢metro bins=x especifica a quantidade de barras ser√£o exibidas no gr√°fico (o n√∫mero depende da estrutura de dados que ser√° utilizado)
```

```python
#o gr√°fico aponta que a maior parte das contas est√£o entre 10 e 20 d√≥lares
```

### jointplot

jointplot() permite que voc√™ combine dois distplots para dados bivariada. com sua escolha de qual **tipo(kind)** par√¢metro voc√™ quer comparar:

* "scatter"
* "reg"
* "resid"
* "kde"
* "hex"

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
#no meio est√° o gr√°fico em scatter
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
# par√¢metro kind para alterar o tipo do gr√°fico a ser exibido
```

```python
#no meio est√° o gr√°fico em hex√°gonos (kind='hex')
#quanto mais escuro o hex√°gono, maior a concentra√ß√£o de pessoas
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
#no meio est√° o gr√°fico em regress√£o (kind='reg')
#√© semelhante ao scatter, por√©m, √© acrescentado a linha de regress√£o

#regress√£o = √© uma ferramenta estat√≠stica utilizada para o monitoramento de processos que possuem interfer√™ncia conjunta de vari√°veis,
#utiliza duas ou mais vari√°veis que apresentam rela√ß√£o de depend√™ncia entre si.

#ser√° mais discutido quando chegar ao t√≥pico de machine learning.
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
#no meio est√° o gr√°fico em densidade (kind='kde')
#o menor circulo √© onde tem mais concentra√ß√£o
```

### pairplot

pairplot vai emparelhar relacionamentos pares em um data frame inteiro (para as colunas numericas) e um argumento de matiz de cores (para colunas categ√≥ricas)

```python
sns.pairplot(tips)
```

```python
#pairplot √© uma boa op√ß√£o para visualizar dados rapidamente
```

```python
sns.pairplot(tips,hue='sex',palette='rocket')
```

```python
#com o parm√¢metro (hue='') √© possivel colorir alguma coluna categ√≥rica
#nesse caso a coluna selecionada foi 'sex' na qual o masculino ficou com uma cor e feminino ficou com outra cor criando um novo destaque
#o par√¢metro palette='' altera o esquema de cores que v√£o ser utilizados na coluna especificada no hue
```

### rugplot

rugplots um conceito muito simples, eles apenas desenham um tra√ßo para cada ponto em uma distribui√ß√£o univariada. Eles s√£o o bloco de constru√ß√£o de um gr√°fico do KDE

```python
sns.rugplot(tips['total_bill'])
```

### kdeplot

kdeplots s√£o [Kernel Density Estimation plots](https://pt.wikipedia.org/wiki/Estimativa_de_densidade_kernel). Esses KDE plots substituem cada observa√ß√£o por uma distribui√ß√£o Gaussiana (Normal) centrada em torno desse valor.

```python
sns.kdeplot(tips['total_bill'])
```
