# Regression Plots

O Seaborn tem muitos recursos integrados para regression plos, no entanto, não discutiremos a regressão até a seção de machine learning, portanto, abordaremos apenas a função **lmplot()** por enquanto.


```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

# lmplot()

### Trabalhando com Markers

lmplot kwargs é passado para **regplot** que é uma forma mais geral de lmplot(). regplot tem um parâmetro scatter_kws que é passado para plt.scatter. Portanto, você deseja definir o parâmetro s nesse dicionário, que corresponde (um pouco confuso) ao tamanho do marcador quadrado. Em outras palavras, você acaba passando um dicionário com os argumentos base matplotlib, neste caso, s para o tamanho de um scatter plot. Em geral, você provavelmente não se lembrará disso de cabeça, mas, em vez disso, consulte a documentação.

```python
# http://matplotlib.org/api/markers_api.html
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'])
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':100})
```

### Usando grid

Podemos adicionar mais separação de variáveis ​​por meio de colunas e linhas com o uso de um grid. Basta indicar isso com os argumentos col ou linha:

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
sns.lmplot(x="total_bill", y="tip", row="sex", col="time",data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm')
```

## Aspect and Size

As figuras Seaborn podem ter seu tamanho e proporção ajustados com os parâmetros **size** e **aspect**:

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',aspect=0.6,height=8)
```
