# Style and Color


Mostramos algumas vezes como controlar a estética da figura em seaborn, mas agora vamos analisá-lo formalmente:

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

## Style


Você pode definir estilos específicos:

```python
sns.set_context('notebook')
#sns.set_context('poster',font_scale=3) estilo de figura
#plt.figure(figsize=(12,3)) tamanho da figura
#sns.set_style('whitegrid') estilo de fundo do grafico
sns.countplot(x='sex',data=tips)
sns.despine() #remove as bordas da direita e topo
#sns.despine(left=True,bottom=True) removendo da esquerda o fundo junto
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='seismic')
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('white')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips,palette='deep')
```

# Spine Removal

```python
sns.countplot(x='sex',data=tips)
sns.despine()
# despine remove a borda superior e da direita, é possivel por parâmetro remover as bordas restantes
```

```python
sns.countplot(x='sex',data=tips)
sns.despine(left=True)
```

## Size e Aspect

Você pode usar o **plt.figure(figsize=(width,height) ** do matplotlib para alterar o tamanho da maioria dos seaborn plots.

Você pode controlar o tamanho e a proporção da maioria dos seaborn plots passando os parâmetros: size e aspect. Por exemplo:

```python
# Non Grid Plot
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

```python
# Grid Type Plot
sns.lmplot(x='total_bill',y='tip',size=2,height=4,data=tips)
```

## Scale e Context

o set_context() permite que você substitua os parâmetros padrão:

```python
sns.set_context('poster',font_scale=4)
sns.countplot(x='sex',data=tips,palette='coolwarm')
```
